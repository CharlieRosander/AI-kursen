```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Docs/data2.csv")
```

```python
df_harmonised = df.copy()
```

```python
df_harmonised
```

```python
def replace_nan_with_mean(column):
    try:
        column = pd.to_numeric(column)
        column.fillna(column.mean(), inplace=True)
    except:
        pass
    return column

df_harmonised = df_harmonised.apply(replace_nan_with_mean)
```

```python
df_harmonised.rename(columns={"age_category": "AGE", "rad_category": "RAD"}, inplace=True)
```

```python
df_harmonised
```

```python
df_harmonised = pd.get_dummies(df_harmonised, columns=["AGE"], prefix = "AGE")
```

```python
df_harmonised = pd.get_dummies(df_harmonised, columns=["RAD"], prefix = "RAD")
```

```python
df_harmonised
```

```python
df_harmonised.to_csv("Docs/data2_harmonised.csv", index=False)
```
